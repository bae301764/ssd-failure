```python
import pandas as pd

MA1 = pd.DataFrame()
for chunk in pd.read_csv("../ssd_data/combined_by_diskmodel/combined_MA1.csv", chunksize=1000000):
    MA1 = pd.concat([MA1, chunk], ignore_index=True)
MA1
```

```python
del(MA1['index'])
```

```python
# null값이 있는 열을 제거하는 코드
def remove_null_columns(df):
    # null인 컬럼을 확인하고, 모든 값이 null인 컬럼을 찾습니다.
    null_columns = df.columns[df.isnull().all()]
    
    # 모든 값이 null인 컬럼을 DataFrame에서 제거합니다.
    df = df.drop(null_columns, axis=1)
    
    return df
```

```python
# 하나의 unique한 값으로만 이루어진 컬럼을 제거하는 코드
def remove_single_value_columns(df):
    single_value_cols = []
    for col in df.columns:
        unique_values = df[col].nunique()
        if unique_values == 1:
            single_value_cols.append(col)

    df_filtered = df.drop(single_value_cols, axis=1)
    return df_filtered
```

```python
MA1 = remove_null_columns(MA1)
MA1 = remove_single_value_columns(MA1)
```

```python
MA1=MA1.fillna(0)
```

```python
# 라벨이 1인 행을 필터링합니다.
selected_disk_model_1 = MA1[MA1['label'] == 1]

# 모델, disk_id, 그리고 날짜 순으로 정렬합니다.
selected_disk_model_1 = selected_disk_model_1.sort_values(by=['disk_id', 'ds'])

# disk_id와 model 열만을 선택하여 새로운 데이터 프레임에 저장합니다.
selected_disk_model_1 = selected_disk_model_1[['disk_id']].drop_duplicates().reset_index(drop=True)

# 결과 데이터 프레임을 출력합니다.
selected_disk_model_1
```

```python
# ds 컬럼에 따라 정렬
MA1 = MA1.sort_values(['disk_id', 'ds'])
# 결과 출력
MA1
```

```python
extracted_disk_failure_model_all = pd.merge(selected_disk_model_1, MA1, on='disk_id', how='inner')
```

```python
# label이 1인 행의 인덱스 찾기
label_index = MA1[extracted_disk_failure_model_all['label'] == 1].index[0]

# 조건에 맞는 행 인덱스 선택
rows_to_label = MA1.iloc[label_index-29:label_index+1]

# disk_id가 변경된 행의 인덱스 찾기
disk_id_changed_index = rows_to_label[rows_to_label['disk_id'].diff() != 0].index

# 변경된 행을 제외하고 label을 1로 변경
rows_to_label.loc[~rows_to_label.index.isin(disk_id_changed_index), 'label'] = 1
```

```python
# label이 1인 행의 인덱스 찾기
label_indices = extracted_disk_failure_model_all[extracted_disk_failure_model_all['label'] == 1].index

# 변경된 값을 적용할 데이터프레임 복사
df_updated = extracted_disk_failure_model_all.copy()

# label이 1인 행을 기준으로 위의 29개 행의 label을 1로 변경
for idx in label_indices:
    start_index = max(idx - 29, 0)
    end_index = idx + 1

    # start_index가 음수인 경우 처리
    if start_index < 0:
        rows_to_label = extracted_disk_failure_model_all.iloc[:end_index]
    else:
        rows_to_label = extracted_disk_failure_model_all.iloc[start_index:end_index]

    # disk_id가 변경된 행이 있는 경우 해당 행을 제외하고 label을 1로 변경
    disk_id_changed_index = rows_to_label[rows_to_label['disk_id'].diff() != 0].index
    rows_to_label.loc[~rows_to_label.index.isin(disk_id_changed_index), 'label'] = 1

    # 변경된 값을 원본 데이터프레임에 적용
    df_updated.update(rows_to_label)
```
