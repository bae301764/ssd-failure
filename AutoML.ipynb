### 5 window 전처리

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# 데이터 프레임 예시 (shuffled_df 라는 데이터 프레임이 있다고 가정)
# shuffled_df = pd.DataFrame(...)

# model과 disk_id별로 unique한 그룹을 찾습니다.
grouped = shuffled_df.groupby(['model', 'disk_id'])

# lag window size
window_size = 5

# 결과를 저장할 빈 데이터프레임을 만듭니다.
result = pd.DataFrame()

# tqdm을 사용하여 진행 상태를 표시합니다.
for name, group in tqdm(grouped, desc="Processing groups"):
    # 변수들에 대한 lag feature를 생성합니다. (4번째부터 마지막에서 두 번째 열까지)
    features = group.iloc[:, 3:-1].values
    labels = group.iloc[:, -1].values
    dates = group.iloc[:, 1].values
    lag_features = []
    lag_labels = []
    lag_dates = []
    
    # window를 이동시키며 lag feature를 만듭니다.
    for i in range(len(features) - window_size + 1):
        window = features[i:i + window_size].flatten()
        lag_features.append(window)
        lag_labels.append(labels[i + window_size - 1])
        lag_dates.append(dates[i + window_size - 1])
    
    # 데이터프레임 형태로 변환합니다.
    num_features = len(group.columns) - 4
    columns = [f'var_{j}_lag_{i}' for j in range(1, num_features+1) for i in range(1, window_size+1)]
    lag_df = pd.DataFrame(lag_features, columns=columns)
    
    # 기존 정보를 포함시킵니다.
    lag_df['model'] = name[0]
    lag_df['disk_id'] = name[1]
    lag_df['date'] = lag_dates  # 날짜를 추가합니다.
    
    # Label을 추가합니다.
    lag_df['label'] = lag_labels
    
    # 결과 데이터프레임에 추가합니다.
    result = pd.concat([result, lag_df])

# 결과를 출력합니다.
print(result)
```

### 30일 이전을 label = 1 로

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# result 데이터프레임 예시
# 'model', 'disk_id', 'date', 'features...', 'label'

# model, disk_id 조합을 가져옴
unique_combinations = result[['model', 'disk_id']].drop_duplicates().values.tolist()

# 결과를 저장할 빈 데이터프레임 생성
processed_data = pd.DataFrame()

# tqdm을 사용하여 진행 상황 표시
for model, disk_id in tqdm(unique_combinations):
    # 특정 model, disk_id에 대한 데이터 추출
    subset = result[(result['model'] == model) & (result['disk_id'] == disk_id)].copy()
    
    # label이 1인 인덱스 찾기
    label_one_indices = subset.index[subset['label'] == 1].tolist()
    
    # label이 1인 날짜 이전 30일을 찾아서 label을 1로 설정
    for index in label_one_indices:
        # 이전 30일의 인덱스를 찾음 (최소 인덱스를 고려)
        start_index = max(index - 30, subset.index.min())
        subset.loc[start_index:index, 'label'] = 1
    
    # 결과 데이터에 추가
    processed_data = pd.concat([processed_data, subset])

# 최종 전처리된 데이터
print(processed_data)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from tqdm import tqdm

# 이전에 만든 lag feature 데이터프레임 예시: processed_data
# 'model', 'disk_id', 'features...', 'label' 컬럼들을 포함

# 각 model, disk_id별로 분리하여 모델 학습 및 예측
unique_combinations = processed_data[['model', 'disk_id']].drop_duplicates().values.tolist()

# 각 조합별 정확도를 저장할 리스트
accuracies = []

# tqdm을 사용하여 진행 상황 표시
for model, disk_id in tqdm(unique_combinations):
    # 특정 model, disk_id에 대한 데이터 추출
    subset = processed_data[(processed_data['model'] == model) & (processed_data['disk_id'] == disk_id)]
    
    # NaN 값을 0으로 대체
    subset = subset.fillna(0)
    
    # 시간 순서를 유지한 상태로 train-test 분할 (예: 앞 80%는 train, 뒤 20%는 test)
    train_size = int(len(subset) * 0.8)
    train_data = subset[:train_size]
    test_data = subset[train_size:]
    
    # 만약 train_data 또는 test_data가 비어있다면, 해당 조합을 건너뜀
    if train_data.empty or test_data.empty:
        continue
    
    # Feature와 label 분리
    X_train = train_data.drop(['model', 'disk_id', 'label'], axis=1)
    y_train = train_data['label']
    X_test = test_data.drop(['model', 'disk_id', 'label'], axis=1)
    y_test = test_data['label']
    
    # RandomForest 모델 학습
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)
    
    # 예측
    y_pred = rf.predict(X_test)
    
    # 정확도 측정 및 저장
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# 전체 평균 정확도
mean_accuracy = sum(accuracies) / len(accuracies) if accuracies else 0
print(f"Mean Accuracy: {mean_accuracy}")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from tqdm import tqdm

# 이전에 만든 lag feature 데이터프레임 예시: processed_data
# 'model', 'disk_id', 'features...', 'label' 컬럼들을 포함

# 각 model, disk_id별로 분리하여 모델 학습 및 예측
unique_combinations = processed_data[['model', 'disk_id']].drop_duplicates().values.tolist()

# 각 조합별 평가 지표를 저장할 리스트
accuracies = []
f1_scores = []
precisions = []
recalls = []
aucs = []

# tqdm을 사용하여 진행 상황 표시
for model, disk_id in tqdm(unique_combinations):
    # 특정 model, disk_id에 대한 데이터 추출
    subset = processed_data[(processed_data['model'] == model) & (processed_data['disk_id'] == disk_id)]
    
    # NaN 값을 0으로 대체
    subset = subset.fillna(0)
    
    # 시간 순서를 유지한 상태로 train-test 분할 (예: 앞 80%는 train, 뒤 20%는 test)
    train_size = int(len(subset) * 0.8)
    train_data = subset[:train_size]
    test_data = subset[train_size:]
    
    # 만약 train_data 또는 test_data가 비어있다면, 해당 조합을 건너뜀
    if train_data.empty or test_data.empty:
        continue
    
    # Feature와 label 분리
    X_train = train_data.drop(['model', 'disk_id', 'label'], axis=1)
    y_train = train_data['label']
    X_test = test_data.drop(['model', 'disk_id', 'label'], axis=1)
    y_test = test_data['label']
    
    # RandomForest 모델 학습
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)
    
    # 예측
    y_pred = rf.predict(X_test)
    
    # 평가 지표 계산 및 저장
    accuracies.append(accuracy_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred, zero_division=0))
    precisions.append(precision_score(y_test, y_pred, zero_division=0))
    recalls.append(recall_score(y_test, y_pred, zero_division=0))
    
    # AUC를 계산하기 전에 테스트 데이터셋에 두 개 이상의 클래스(라벨)가 있는지 확인
    if len(set(y_test)) > 1 and hasattr(rf, "predict_proba"):
        y_pred_proba = rf.predict_proba(X_test)
        if y_pred_proba.shape[1] > 1:
            aucs.append(roc_auc_score(y_test, y_pred_proba[:, 1]))
        else:
            aucs.append(float('nan'))  # AUC를 계산할 수 없는 경우 NaN을 추가
    else:
        aucs.append(float('nan'))  # AUC를 계산할 수 없는 경우 NaN을 추가
```

```python
# 전체 평균 평가 지표
mean_accuracy = sum(accuracies) / len(accuracies) if accuracies else 0
mean_f1_score = sum(f1_scores) / len(f1_scores) if f1_scores else 0
mean_precision = sum(precisions) / len(precisions) if precisions else 0
mean_recall = sum(recalls) / len(recalls) if recalls else 0
mean_auc = sum(aucs) / len(aucs) if aucs else 0

# 결과 출력
print(f"Mean Accuracy: {mean_accuracy}")
print(f"Mean F1 Score: {mean_f1_score}")
print(f"Mean Precision: {mean_precision}")
print(f"Mean Recall: {mean_recall}")
print(f"Mean AUC: {mean_auc}")
```
