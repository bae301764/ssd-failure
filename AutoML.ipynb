```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
df1 = pd.DataFrame()
for chunk in pd.read_csv("../combined_data/combined_df01.csv", chunksize=1000000):
    df1 = pd.concat([df1, chunk], ignore_index=True)
df1
```

```python
df2 = pd.DataFrame()
for chunk in pd.read_csv("../combined_data/combined_df02.csv", chunksize=1000000):
    df2 = pd.concat([df2, chunk], ignore_index=True)
df2
```

```python
df = pd.concat([df1,df2], ignore_index=True)
```

```python
# 라벨이 1인 행을 필터링합니다.
filtered_df = df[df['label'] == 1]

# disk_id와 model 열만을 선택하여 새로운 데이터 프레임에 저장합니다.
result_df = filtered_df[['disk_id', 'model']].drop_duplicates().reset_index(drop=True)

# 결과 데이터 프레임을 출력합니다.
print(result_df)
```

```python
# result_df에 있는 disk_id와 model 조합을 사용합니다.
selected_disk_model = result_df.values.tolist()

# 결과를 저장할 빈 데이터프레임 생성
final_df = pd.DataFrame()

# 각 disk_id와 model에 대해 원본 데이터프레임에서 행을 필터링합니다.
for disk_id, model in selected_disk_model:
    filtered_rows = df[(df['disk_id'] == disk_id) & (df['model'] == model)]
    final_df = pd.concat([final_df, filtered_rows])

# 모델, disk_id, 그리고 날짜 순으로 정렬합니다.
final_df = final_df.sort_values(by=['model', 'disk_id', 'ds'])

# 최종 결과를 출력합니다.
print(final_df)
```

```python
import random

# 시드 값을 설정합니다.
random.seed(42)

# label이 0인 disk_id와 model 조합 추출
label_0_df = df[df['label'] == 0][['disk_id', 'model']].drop_duplicates().reset_index(drop=True)

# 814개를 랜덤하게 선택
selected_disk_model_0 = random.sample(label_0_df.values.tolist(), 814)
```

```python
from tqdm import tqdm

# 결과를 저장할 빈 데이터프레임 생성
final_df2 = pd.DataFrame()

# 진행 상황을 표시하기 위해 tqdm 사용
for disk_id, model in tqdm(selected_disk_model_0):
    filtered_rows = df[(df['disk_id'] == disk_id) & (df['model'] == model)]
    final_df2 = pd.concat([final_df2, filtered_rows])

# 모델, disk_id, 그리고 날짜 순으로 정렬합니다.
final_df2 = final_df2.sort_values(by=['model', 'disk_id', 'ds'])
```

```python
# final_df와 final_df2를 합칩니다.
combined_df = pd.concat([final_df, final_df2])

# model, disk_id, date를 기준으로 그룹화하고 shuffle합니다.
grouped = combined_df.groupby(['model', 'disk_id'])
shuffled_groups = random.sample(list(grouped), len(grouped))
shuffled_df = pd.concat([group for _, group in shuffled_groups])

# 최종 결과를 출력합니다.
print(shuffled_df)
```

```python
# final_df와 final_df2를 합칩니다.
combined_df = pd.concat([final_df, final_df2])

# model, disk_id, date를 기준으로 그룹화하고 shuffle합니다.
grouped = combined_df.groupby(['model', 'disk_id'])
shuffled_groups = random.sample(list(grouped), len(grouped))

# 진행 상황을 표시하기 위해 tqdm 사용
shuffled_df = pd.DataFrame()
for _, group in tqdm(shuffled_groups):
    shuffled_df = pd.concat([shuffled_df, group])

# 최종 결과를 출력합니다.
print(shuffled_df)
```

```python
shuffled_df.to_csv('../combined_data/combined_subset.csv',index=False)
```
